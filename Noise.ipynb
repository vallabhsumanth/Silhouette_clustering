# Gaussian Filters

```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
```

```python
# Step 2: Load the digits dataset
digits = load_digits()
images = digits.images  # This contains the 8x8 pixel images
labels = digits.target  # Labels for the digits
```

```python
# Step 3: Add Gaussian noise to the images
# We use a noise factor (0.3) to control the level of noise added to the images
noise_factor = 0.3
noisy_images = images + noise_factor * np.random.randn(*images.shape)

# Step 4: Clip values to ensure they stay within the valid range of pixel values (0 to 16)
noisy_images = np.clip(noisy_images, 0, 16)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(-10, 10, 1000)

# Define Gaussian functions
def gaussian(x, mean, std_dev):
    return (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev)**2)

# Parameters for the Gaussians
mean1, std_dev1 = 0, 1  # Standard normal distribution
mean2, std_dev2 = 0, 2  # Wider spread
mean3, std_dev3 = 2, 1  # Shifted mean

# Compute Gaussian values
y1 = gaussian(x, mean1, std_dev1)
y2 = gaussian(x, mean2, std_dev2)
y3 = gaussian(x, mean3, std_dev3)

# Plot the Gaussian curves
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label="Mean=0, Std Dev=1 (Tall & Narrow)", color='blue')
plt.plot(x, y2, label="Mean=0, Std Dev=2 (Wide & Short)", color='green')
plt.plot(x, y3, label="Mean=2, Std Dev=1 (Shifted Right)", color='red')

# Add labels and legend
plt.title("Gaussian Distribution with Different Means and Standard Deviations", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("Probability Density", fontsize=12)
plt.axvline(x=0, color='gray', linestyle='--', label='Mean=0')
plt.axvline(x=2, color='red', linestyle='--', label='Mean=2')
plt.legend(fontsize=10)
plt.grid(alpha=0.3)
plt.show()
```

```python
# Step 5: Visualize one of the noisy images
plt.imshow(noisy_images[0], cmap='gray')
plt.title(f"Digit: {labels[0]} with Noise")
plt.show()
```

# Denosizing The Data

```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
from scipy.ndimage import gaussian_filter
```

```python

digits = load_digits()
images = digits.images  
labels = digits.target  
```

```python

noise_factor = 0.3
noisy_images = images + noise_factor * np.random.randn(*images.shape)

# Clip values to ensure they stay in the valid range of pixel values (0 to 16)
noisy_images = np.clip(noisy_images, 0, 16)
```

```python

sigma = 1 
denoised_images = gaussian_filter(noisy_images, sigma=sigma)

# Clip values again after applying the filter (to ensure they stay in the valid range)
denoised_images = np.clip(denoised_images, 0, 16)
```

```python
# Step 5: Visualize the original noisy image and denoised image
plt.figure(figsize=(8, 4))

# Plot the noisy image
plt.subplot(1, 2, 1)
plt.imshow(noisy_images[0], cmap='gray')
plt.title(f"Noisy Image: {labels[0]}")

# Plot the denoised image
plt.subplot(1, 2, 2)
plt.imshow(denoised_images[0], cmap='gray')
plt.title(f"Denoised Image: {labels[0]}")

plt.show()
```

# Median Filter

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from scipy.ndimage import median_filter
```

```python
# Load the digits dataset from scikit-learn
digits = load_digits()
images = digits.images  # The dataset contains 8x8 grayscale images of digits
```

```python
# Select an example image (first image, which represents the digit '0')
image = images[0]
```

```python
# Apply the median filter with a 3x3 kernel
filtered_image = median_filter(image, size=3)
```

```python
# Plot the original and filtered images side-by-side
plt.figure(figsize=(10, 5))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')

# Filtered Image
plt.subplot(1, 2, 2)
plt.imshow(filtered_image, cmap='gray')
plt.title("Filtered Image (Median)")
plt.axis('off')

plt.tight_layout()
plt.show()
```
